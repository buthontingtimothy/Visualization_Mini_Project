# Import Necessary Library

```python
# data maniuplation
import pandas as pd
```

# Read CSV of Data and GCS (Geographic Coordinate System)

```python
# read csv for data series and gcs, ignore addition index column by setting index_col = 0
data = pd.read_csv("data_final.csv", index_col=0)
gcs = pd.read_csv("gcs.csv", index_col=0)
```

```python
# merge two dataframe by region as key
combine = data.merge(gcs, on='region', how='left')
```

```python
combine
```

# Save as csv

```python
# save it in csv
combine.to_csv("data_gcs_combine.csv")
```
