# Import Necessary Library

```python
import pandas as pd
```

```python
combine = pd.read_csv("data_gcs_combine.csv", index_col=0)
```

# Fill Null for New Region from Old Region

```python
def region_update(df, old, new):
    df = df.copy()
    df_new = df[df['region'] == new]
    df_old = df[df['region'] == old]
    df_new_reset_index = df_new.reset_index()
    df_old_reset_index = df_old.reset_index()
    update = df_new_reset_index.combine_first(df_old_reset_index).set_index('index')
    df_new_index = df_new.index.tolist()
    df_old_index = df_old.index.tolist()
    list_drop = df_new_index + df_old_index
    df_drop = df.drop(list_drop)
    df_combine = pd.concat([df_drop, update])
    df_sort = df_combine.sort_values(['date','region']).reset_index(drop=True)
    return df_sort
```

```python
region_old = "Tai Po(Conservation Studies Centre)"
region_new = "Tai Po(Yuen Chau Tsai Park)"
```

```python
combine_update = region_update(combine, region_old, region_new)
```
